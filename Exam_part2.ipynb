```python
df.groupby('Product Type')['Quantity'].sum()
#here im showing the total number of sales per product type
```

```python
#to see this previous information in a more visual way im creating a bar chart of the number of sales per product.
#by analyzing the chart we can concluse that smartphones are the most sold products and headphones the lest sold.
df_sales_product = df.groupby('Product Type')['Quantity'].sum().reset_index() #This line calculates the total quantity sold for each 'Product Type'.

plt.figure(figsize=(10, 6)) #this creates a new figure for the plot with a specified size of 10 inches wide and 6 inches tall.
sns.barplot(x='Product Type', y='Quantity', data=df_sales_product) #the nexts lines are used to create the barplot and the details of the chart
plt.title('Number of Sales per Product Type')
plt.xlabel('Product Type')
plt.ylabel('Number of Sales')
plt.show()
```

```python
#Here im analyzing the average rating per product type
df_rating_product = df.groupby('Product Type')['Rating'].mean().reset_index()

plt.figure(figsize=(10, 6)) #this will create a bar char with seaborn to visualize the average rating per product.
sns.barplot(x='Product Type', y='Rating', data=df_rating_product) #in the following lines i indicate the specificatiosn for the graph
plt.title('Average Rating per Product Type')
plt.xlabel('Product Type')
plt.ylabel('Average Rating')
plt.ylim(0, 5)
plt.show()

#By analyzing this bar char we can conclude that in general the products are rated around 3 (as we saw before in the total mean of the rating),
#but we can also see that smartphones are the best rated product.
```

```python
#Next thing is creating a histogram for total price to see the shape of the data
plt.figure(figsize=(10, 6)) #this graph as the previous ones is following the same steps. First defining the size and then the specificatiosn of the graph
sns.histplot(df['Total Price'], kde=True)
plt.title('Histogram of Total Price')
plt.xlabel('Total Price')
plt.ylabel('Frequency')
plt.show()

#In conclusion the data is right skewed showing the majority of the data around 0 and 5000.
#This means that there are only a few products that are more expensive than this, which is really normal in datasets of business purchases.
```

```python
#Now i wanted to analyze if there is a correlation between the raiting and total price of the product.
plt.figure(figsize=(10, 6)) #Same as the previous graphs im adding the size and all the specifications that i want to see
sns.scatterplot(x=df['Rating'], y=df['Total Price'])
plt.title('Scatter Plot of Rating vs Total Price')
plt.xlabel('Rating')
plt.ylabel('Total Price')
plt.show()

#In conclusion, by looking at the graph i cannot recognize any real correlation between those 2 variables.
```

```python
# prompt: scatter plot quantity and unit price
#I created another scatter plot to see if there was a correlation between quantity and unit price, but again the 2 variables do not seem strongly correlated, as we can see by how spread the points are.
plt.figure(figsize=(10, 6)) #Same as the previous graphs im adding the size and all the specifications that i want to see
sns.scatterplot(x=df['Quantity'], y=df['Unit Price'])
plt.title('Scatter Plot of Quantity vs Unit Price')
plt.xlabel('Quantity')
plt.ylabel('Unit Price')
plt.show()
```

```python
#I created the boxplot for total price by product.
plt.figure(figsize=(10, 6)) #Same as the previous graphs im adding the size and all the specifications that i want to see
sns.boxplot(x='Product Type', y='Total Price', data=df) #defining the axis and where is the data for teh graph
plt.title('Boxplot of Total Price by Product Type')
plt.xlabel('Product Type')
plt.ylabel('Total Price')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

#In concluision there's a clear hierarchy in the value of individual transactions: purchases of smartphones and tablet are typically of higher value, while those of headphones are at the lower end.
#The chart is a comparison of the distribution of typical prices and the dispersion of spending for each item type, providing insights into which products drive the majority of each transaction's value.
```

```python
#Now i wanted to see if there is a difference in shipping type per product type
plt.figure(figsize=(12, 7)) #Same as the previous graphs im adding the size and all the specifications that i want to see
sns.countplot(data=df, x='Product Type', hue='Shipping Type')
plt.title('Shipping Type Distribution by Product Type')
plt.xlabel('Product Type')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Shipping Type')
plt.tight_layout()
plt.show()

#The plot shows that most product types have a similar distribution across shipping types.
#However, smartphones appear to have higher proportion of standard, overnight and express shipping compared to the other categories, which might indicate that customers buying these more expensive items are more willing to pay for faster shipping.
```
